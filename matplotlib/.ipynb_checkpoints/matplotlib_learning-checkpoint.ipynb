### Pyplot API and Object-Oriented (OO) API
- Pylot -> quickly
- OO -> advance

```python
x = np.linspace(0,10,100)
plt.plot(x,np.sin(x));
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.xlim([0,2*np.pi]);
plt.ylim([-1,1]);
```

```python
x = np.linspace(0,10,100)
plt.plot(x,np.sin(x), linestyle="--", label="sin(x)");
plt.plot(x,np.cos(x), label="cos(x)");

plt.title("sin and cos")
plt.xlabel("x")
plt.ylabel("f(x)")

# plt.xlim([0,2*np.pi]);
# plt.ylim([-1,1]);

# plt.axis([xmin, xmax, ymin, ymax])
plt.axis([0, 2*np.pi, -1, 1])

plt.legend();
```

```python
x = np.linspace(0,10,100)
plt.plot(x,np.sin(x), linestyle="--", label="sin(x)");
plt.plot(x,np.cos(x), label="cos(x)");

plt.title("sin and cos")
plt.xlabel("x")
plt.ylabel("f(x)")

plt.axis("tight")

plt.legend();
```

```python
x = np.linspace(0,10,100)
plt.plot(x,np.sin(x), linestyle="--", label="sin(x)");
plt.plot(x,np.cos(x), label="cos(x)");

plt.title("sin and cos")
plt.xlabel("x")
plt.ylabel("f(x)")

plt.axis("equal")

plt.legend();
# plt.show();
```
