```python
df["total_price"] = df["item_price"] * df["quantity"]
# thêm 1 cột total_price bằng blablabla vào dataframe
```

```python
df.total_price.sum()
# Trả về tổng tất cả các giá trị trong cột
```

```python
# Group by
df.groupby("item_name")
```

```python
# Xuất ra
# Và chỉ lấy cột quantity thôi
df.groupby("item_name")["quantity"].apply(print)
```

```python
# Lấy tổng lại
temp = df.groupby("item_name")["quantity"].sum()
temp.sort_values()
# sort theo value
# mặc định là tăng dần
```

```python
temp = df.groupby("item_name")["quantity"].sum()
temp.sort_values(ascending=False)
```

```python
# Lấy 5 giá trị đầu
temp = df.groupby("item_name")["quantity"].sum()
temp.sort_values(ascending=False).head()
```
